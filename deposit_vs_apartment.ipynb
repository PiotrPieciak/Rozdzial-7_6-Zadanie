####Import niezbędnych bibliotek

```python
import numpy as np
import numpy_financial as npf
import matplotlib.pyplot as plt
```

####Wpisujemy podstawowe dane dla naszego zadania

```python
pv=120000
years=5
rate_apartment = 0.05
rate_deposit = 0.12
freq=12
```

####Przeprowadzamy wstępne kalkulacje

```python
rate_deposit /= freq
nper = years*freq
```

####Obliczamy ile bedzie wynosiła orientacyjna cena mieszkania za 5lat

```python
final_apartment_cost = np.linspace(pv, -np.around(npf.fv(rate_apartment, years, 0 , pv)),nper)
```

```python
print("Orientacyjna cena mieszkania po 5 latach będzie wynosiła: ",final_apartment_cost[-1],"zł")
```

####Obliczenie Ile trzeba wpłacać do banku każdego miesiąca, aby uzbierać na mieszkanie w ciągu 5 lat

```python
monthly_savings = - np.around(npf.pmt( rate_deposit, nper , 0, final_apartment_cost[-1]))
```

```python
print("Każdego miesiąca należy wpłacać ",monthly_savings,"zł, żeby uzbierać na mieszkanie w ciągu 5 lat.")
```

#####npf.fv(0.12/12, 5*12, 1875, 0) # Weryfikacje oblicznonej wartości można wykonać kodem po lewej 

####Tworzymy listę do wykresu która bedzie zawierać kolejne wartości depozytu w banku dla kolejnych miesiecy

```python
deposit_value = - np.around(npf.fv(rate_deposit, np.arange(1,61) , monthly_savings, 0))
```

####Stworzenie wykresu wymaganego w pkt 3 zadania

```python
plt.plot(final_apartment_cost,label='cena mieszkania')
plt.plot(deposit_value, label='wartość lokaty')
plt.legend()
plt.title('Porównanie skumulowanej wartości mieszkania i lokaty dla okresy 5lat')
plt.xlabel('Liczba okresów')
plt.ylabel('Skumulowana wartość')
```
